```python
#loads json

Months = {"Dec":12,"Nov":11,"Oct":10,"Sep":9,"Aug":8,"Jul":7,"Jun":6,"May":5,"Apr":4,"Mar":3,"Feb":2,"Jan":1}

#get the xml text by id

def get_poll_json(ids):
    xml_url = r'http://www.realclearpolitics.com/epolls/json/{ids}_historical.js'.format(ids=ids)
    xml_text = requests.get(xml_url).text.replace('return_json(','')[:-2]
    return xml_text


def get_json_content(text):
    return json.loads(text)["poll"]["rcp_avg"]


def convert_date(date):
    date_items = date.split(' ')
    day = date_items[1]
    month = str(Months[date_items[2]])
    year = date_items[3]
    return "{month}/{day}/{year}".format(month=month,
                                          day = day,
                                          year = year)
    
def rcp_poll_data(json_file):
    date_list = []
    obama_list = []
    romney_list = []
    for section in json_file:
        date_info = convert_date(section["date"])
        obama_value = section["candidate"][0]["value"]
        romney_value = section["candidate"][1]["value"]
        date_list.append(date_info)
        obama_list.append(float(obama_value))
        romney_list.append(float(romney_value))
    
    #print obama_list
    return pd.DataFrame({'Obama' :np.array(obama_list), 
                         'Romney':np.array(romney_list)},index = pd.to_datetime(date_list))



def poll_plot(poll_id):
    url_text = get_poll_json(1171)
    json_file = get_json_content(url_text)
    pd_data = rcp_poll_data(json_file)

    plt.plot(pd_data.index,pd_data.Obama,label = "Obama")
    plt.plot(pd_data.index,pd_data.Romney,label = "Romney")
    
    plt.xticks(rotation=45)
    plt.legend(loc='best')
    plt.xlabel("Obama V.S Romney")
    plt.ylabel("Normalized Poll Percentage")
    
poll_plot(1171)


    
```
